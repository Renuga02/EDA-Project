```python
#Exploratory Data Analysis:
```

```python
# plotting a bar chart for Gender and it's count
a=sns.countplot(x='Gender',data=df)

for bars in a.containers:
    a.bar_label(bars)
plt.title("Bar plot")    
```

```python
df['Amount'].sum()
```

```python
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen
```

```python
sales_gen
```

```python
sns.barplot(x='Gender',y='Amount',data=sales_gen)
```

```python
#From above graphs we can see that most of the buyers are females and even the purchasing power of females are greater than men
```

```python
#Age:
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')

for bars in ax.containers:
    ax.bar_label(bars)


```

```python
# Total Amount vs Age Group:
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_age
```

```python
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

```python
#From above graphs we can see that most of the buyers are of age group between 26-35 yrs female
```

```python
#State:
```

```python
# total number of orders from top 10 states
```

```python
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sales_state
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
# total amount/sales from top 10 states
```

```python
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

```python
#From above graphs we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively
```

```python
#Marital Status:
```

```python
ax=sns.countplot(x='Marital_Status',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_status=df.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
```

```python
sns.barplot(data=sales_status,x='Marital_Status',y='Amount')
```

```python
#From above graphs we can see that most of the buyers are married (women) and they have high purchasing powe
```

```python
#Occupation:
```

```python
sns.set(rc={'figure.figsize' : (21,5)})
ax=sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_OCC=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_OCC
```

```python
sales_OCC=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(21,5)})
sns.barplot(data=sales_OCC,x='Occupation',y='Amount')
```

```python
#From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector
```

```python
#Product Category:
```

```python
sns.set(rc={'figure.figsize' :(21,5)})
ax=sns.countplot( data =df,x='Product_Category')
for bars in ax.containers:
     ax.bar_label(bars)
```

```python
sales_pro=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize' : (22,5)})
sns.barplot(x='Product_Category', y='Amount',data=sales_pro)
```

```python
#From above graphs we can see that most of the sold products are from Food, Clothing and Electronics category
```

```python
sales_state=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
```

```python
sns.set(rc={'figure.figsize' :(20,5)})
sns.barplot(x='Product_ID',y='Orders',data=sales_state)
```

```python
# top 10 most sold products (same thing as above)
```
